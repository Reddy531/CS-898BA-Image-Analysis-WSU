```python
import cv2
import matplotlib.pyplot as plt
import numpy
```

```python
img1 = cv2.imread('road_image.jpg')
img2 = cv2.imread('00-puppy.jpg')
```

```python
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
f = plt.figure(figsize=(12,12))
f.add_subplot(1,2, 1)
plt.imshow(img1)
f.add_subplot(1,2, 2)
plt.imshow(img2)
plt.show(block=True)
```

```python
img1 = cv2.imread('drawing_1.png')
img2 = cv2.imread('drawing_2.png')


bit_and = cv2.bitwise_and(img2, img1)
bit_or = cv2.bitwise_or(img2, img1)
bit_xor = cv2.bitwise_xor(img1, img2)
bit_not = cv2.bitwise_not(img1)
bit_not2 = cv2.bitwise_not(img2)
```

```python
f = plt.figure(figsize=(20,20))

f.add_subplot(1,2, 1)
plt.imshow(img1)
f.add_subplot(1,2, 2)
plt.imshow(img2)
```

```python
f = plt.figure(figsize=(20,20))

f.add_subplot(1,5, 1)
plt.imshow(bit_and)
f.add_subplot(1,5, 2)
plt.imshow(bit_or)
f.add_subplot(1,5, 3)
plt.imshow(bit_xor)
f.add_subplot(1,5, 4)
plt.imshow(bit_not)
f.add_subplot(1,5, 5)
plt.imshow(bit_not2)
plt.show(block=True)
```

```python
img2 = cv2.imread('00-puppy.jpg')

img_bw = cv2.cvtColor(img2,cv2.COLOR_RGB2GRAY)
plt.imshow(img_bw,cmap='gray')
```

```python
def inverte(imagem):
    imagem = (255-imagem)
    return imagem
```

```python
inv_img_bw = inverte(img_bw)
plt.imshow(inv_img_bw,cmap='gray')
```

```python
hist_full = cv2.calcHist([inv_img_bw],[0],None,[256],[0,256])
```

```python
plt.plot(hist_full)
```

```python
import numpy as np
img = cv2.imread('00-puppy.jpg',0)
equ = cv2.equalizeHist(img)
res = np.hstack((img,equ)) #stacking images side-by-side
cv2.imwrite('res.png',res)
```

```python
hist_full = cv2.calcHist([equ],[0],None,[256],[0,256])
plt.plot(hist_full)
```

```python
img = cv2.imread('00-puppy.jpg',0)
rows,cols = img.shape

M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))
plt.imshow(dst,cmap='gray')
```

```python
M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.imshow(dst,cmap='gray')
```
